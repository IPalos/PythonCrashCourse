### Celcius <-> Fahrenheit

El objetivo de este ejercicio es crear un script capaz de convertir la temperatura de Celcius a Fahrenheit.
El usuario ingresará un String de la forma `{valor}C` si es Celcius, o `{valor}F` si es Fahrenheit ( ejemplo: `"43C"`,`"102F`). Y el usuario espera recibir el valor ya convertido, en el mismo formato que ingresó el dato.

**Objetivos:**
- Recibir el input del usuario como un String ( ejemplo: `"43C"`,`"102F`)
- Checar si es Celcius o Fahrenheit
- Convertir el valor
- Imprimir el valor convertido

### Solución

```python
valor_a_convertir=input("Ingresa la temperatura: ")             # Recibe el valor del usuario
if valor_a_convertir[-1].lower() in {"c","f"}:                # Verifica si el último caractér de valor_a_convertir sea c o f
    if valor_a_convertir[-1].lower() == "c":                    # Verifica si es C
        resultado= (int(valor_a_convertir[0:-1])*9/5) + 32    # Convierte el valor con la formula C = (F*9/5)+32
        print("Resultado: "+str(resultado)+"°F")              # Imprime el resultado
    else:                                                     #si no es C, entonces es F
        resultado = (int(valor_a_convertir[0:-1])-32)*5/9     # Convierte el valor con la formula F = (C-32)*5/9
        print("Resultado: "+str(resultado)+"°C")              # Imprime el resultado
else:
    print("No ingresaste un valor válido")                     # Imprime el error
```

### Código de barras válido?

El código ISBN (International Standard Book Number) es un código de 10 dígitos que sirve para identificar cada libro.
y se ven de la forma :


Donde el primer dígito indica el país de origen del libro, según la siguiente tabla:

| Código | Pais |
|---|---|
| 0 | UK |
| 1 | US |
| 2 | France |
| 3 | Germany |
| 4 | Japan |
| 5 | Russia |

También, para verificar si el código de barras fue leído correctamente por la computadora, la suma de todos los dígitos del código es múltiplo de 11.


**Objetivos:**
- Recibir un string que contiene puros números, y verificar que contiene 10 dígitos
- Verificar si el código ingresado viene de un país válido, y decir de qué país es
- Verificar si la suma de sus números es multiplo de 11 
- Imprimir el país de origen, y si el código es válido o no 
 
 **Ejemplos:**
 
 
| Input de usuario | Output del Script |
|---|---|
| 0330284987 | (UK, si funciona) |
| 1581820089 | (US, no funciona) |
| 2226052572 | (France, si funciona) |
| 3796519008 | (Germany, no funciona) |

### Solución

```python
barcode = input("Ingresa el código ISBN: ")

#Guarda los códigos de cada país en el índice adecuado
countryCodes = ["UK", "US", "Francia", "Alemania", "Japón", "Rusia"]

# Verifica si el código de barras tiene 10 dígitos
if len(barcode) == 10:
    
    #Verifica si el primer dígito está en el intervalo (0,5)
    if int(barcode[0]) in range(0,6):
        
        #Almacena el nombre del país en country, según el primer elemento del código
        country = countryCodes[int(barcode[0])]
        
        #Suma el valor de todos los dígitos, y los guarda en la variable "total"
        total=0
        for digit in barcode:
            total += int(digit)
        
        #Verifica que "total" es multiplo de 11
        if total%11 == 0:
            print("Tu libro viene de: "+country+". Su código es VÁLIDO.")
        else:
            print("Tu libro viene de: "+country+". Su código es INVÁLIDO.")
        
    else:
        print("Tu código no empieza con un valor válido")
else:
    print("Tu código no tiene 10 dígitos")
```

### Adivinar que numero piensa la compu

Hagamos un script que haga que la computadora "piense" un número, del 0 al 100, y que el usuario tenga 6 intentos para adivinar ese número. Cuando el usuario hace un intento, la computadora debe decir si el número adivinado es el correcto, y en caso de no serlo, debe decir si es mayor o menor.

**Objetivos:**
- Generar un número aleatorio entre el 0 y el 100
- Darle 6 intentos al jugador
- Recibir el input del jugador
- Verificar si es correcto
- En caso de no serlo, decirle si el número pensado fue mayor o menor.


### Solución

```python
#Importar la librería random
import random

#Generar un número aleatorio entre el 0 y el 100
computersNum = random.randint(0,100)

#Darle 6 intentos al jugador
numOfGuesses = 6

#Corre todo el bloque interno mientras tengas intentos
while numOfGuesses > 0:

    #Recibir el input del jugador
    guess = input("Adivina cual es el número que estoy pensando: ")

    #Verificar si el intento es correcto
    if int(guess) == computersNum:
        print("Felicidades! Adivinaste mi número!")
        break

    #En caso de no serlo, decirle que es menor
    elif int(guess) > computersNum:
        print("El numero que estoy pensando es menor a "+guess)
        numOfGuesses-=1

    #O mayor
    else:
        print ("El numero que estoy pensando es mayor a "+guess)
        numOfGuesses-=1
```
