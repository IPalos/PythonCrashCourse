## Generadores de datos

```python
import random

# Genera "size" tiradas de un dado
def generate1dieRoll(size,dieSize):
    res=[]
    for i in range(size):
        die=random.randint(1,dieSize)
        res.append(die)
    return res

# Genera "size" tiradas de 2 dados
def generate2dieRoll(size,dieSize):
    res=[]
    for i in range(size):
        die1=random.randint(1,dieSize)
        die2=random.randint(1,dieSize)
        res.append(die1+die2)
    return res


#Genera "size" tiradas de 3 dados, y luego le quita la tirada menor
def generate3_2dieRoll(size, dieSize):
    res=[]
    for i in range(size):
        die1=random.randint(1,dieSize)
        die2=random.randint(1,dieSize)
        die3=random.randint(1,dieSize)
        res.append(die1+die2+die3-min(die1,die2,die3))
    return res


# Genera una caminata aleatoria iniciando en startPoint, con pasos step
def randomWalker(startpoint, step, size):
    res=[]
    currentPoint=startpoint
    for i in range(size):
        currentPoint+= random.choice([-step,step])
        res.append(currentPoint)
    return res     
    
```

```python
%matplotlib inline
plt.plot(randomWalker(0,1,1000))
```

```python
%matplotlib inline
plt.hist(generate3_2dieRoll(50000,6))
plt.hist(generate2dieRoll(50000,6))
```
