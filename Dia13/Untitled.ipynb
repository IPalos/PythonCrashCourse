```python
import requests as req
import pandas as pd
```

```python
def generarURL():
    campos = ['ano_inicio','mes_inicio','fechainicio','delito',
 'categoria','sexo','tipopersona','calidadjuridica','competencia',
'ano_hecho','mes_hecho','fechahecho']
    #el url de la api sin ninguna  petición
    url_inicial="https://datos.cdmx.gob.mx/api/records/1.0/search/?dataset=victimas-en-carpetas-de-investigacion-pgj"
    #cuantos delitos se van a pedir
    filas = input("Cuantos delitos quiere obtener? ")
    url_inicial+="&rows="+filas
    
    #los campos que se quieren obtener
    for campo in campos:
        pregunta = input("Quiere consultar el campo: "+campo+"?(y/n) ")
        if pregunta.lower() == "y":
            url_inicial+="&facet="+campo
        
    for filtro in campos:
        pregunta = input("Quiere agregar el filtro de: "+filtro+"?(y/n) ")
        if pregunta.lower() == "y":
            valor_filtro = input("Que valor quiere que tome el filtro?")
            url_inicial+="&refine."+filtro+"="+valor_filtro
            
            
    return url_inicial
    
    
    
    
    
```

```python
url = generarURL()
```

```python
datosApi= req.get(url).json()
datosApi
```

```python
tabla1 = pd.DataFrame(datosApi["records"])

data = [item.values() for item in tabla1["fields"]]

tabla2=pd.DataFrame(data,columns= tabla1["fields"][0].keys())



tabla2

```
