### Dibujar la diagonal de un cuadrado, del tamaño que pida el usuario
Entendamos un poco cómo funciona Paint( el programa básico de dibujo de Windows) cuando le pedimos que nos dibuje una línea recta que va del extremo a otro de un cuadrado de **n** pixeles. Básicamente lo que sucede es que guarda el color (para este ejercicio, digamos que 0= blanco y 1=negro) de cada pixel en una *matriz* del tamaño de la pantalla ,por ejemplo, muchas pantallas tienen 1920x1080 pixeles. Ahora tomemos una región cuadrada de esa pantalla, y queremos dibujarle una diagonal.Para eso, hay que preguntarle al usuario de qué tamaño es el cuadrado. Una vez que el usuario haya ingresado el tamaño del cuadrado ( que se puede ver como una lista de **n** listas, a su vez de **n** enteros, ya sea 0 o 1) ahora hay que distinguir entre los elementos que están en la diagonal y los que están fuera de ella. **HINT**:Los elementos de la diagonal tienen la propiedad de que su posición en la fila es la misma que su posición en la columna.

Este problema también se puede asociar a la colocación de mosaicos en un cuadrado, donde la diagonal tiene mosaicos de distinto color
